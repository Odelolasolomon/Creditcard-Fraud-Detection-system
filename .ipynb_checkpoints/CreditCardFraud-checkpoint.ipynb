### Feature Engineering
- In this part of this project. Relevant featutres will be engineered.

```python
df.head() 
```

```python
df['ratio_distancehome_transaction']=df['distance_from_home']/df['distance_from_last_transaction']
df['ratio_distancehome_medianprice']=df['distance_from_home']/df['ratio_to_median_purchase_price']
```

```python
df.head()
```

```python
df['difference_in_home_last_trans']=df['ratio_distancehome_transaction']-df['distance_from_last_transaction']
```

```python
df.dtypes  
```

```python
x=df.drop(columns=['fraud'],axis=1)
y=df['fraud']  
```

```python
smote= SMOTE()
x_resampled, y_resampled= smote.fit_resample(x,y) 
```

```python
plt.figure(figsize=[5,4])
data= pd.concat([x_resampled, y_resampled],axis=1)
sns.countplot(data=data, x='fraud')
plt.title('Count of fraud');
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.20, shuffle=True, random_state=True) 
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report 

#Logistic Regression
logreg=LogisticRegression()
logreg.fit(x_train_scaled, y_train)
y_logreg=logreg.predict(x_test)
y_logreg_prob=logreg.predict_proba(x_test)[:,1]
```

```python
#Performance metrics evaluation
print("Accuracy:\n",accuracy_score(y_test,y_logreg))
print("Precision:\n",classification_report(y_test,y_logreg))
```

```python
#plotting the ROC curve
from sklearn.metrics import auc
fpr,tpr,thresholds=metrics.roc_curve(y_test,y_logreg_prob)
plt.plot(fpr,tpr)
plt.plot([0,1],[0,1],'r-.')
plt.xlim([-0.2,1.2])
plt.ylim([-0.2,1.2])
plt.title('Receiver Operating Characteristic\nLogistic Regression')
plt.legend(loc='lower right')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report 

#Logistic Regression
tree=DecisionTreeRegressor()
tree.fit(x_train_scaled, y_train)
y_tree=tree.predict(x_test)
```

```python
#Performance metrics evaluation
print("Accuracy:\n",accuracy_score(y_test,y_tree))
print("Precision:\n",classification_report(y_test,y_tree))
```

### Random forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report 

#Logistic Regression
forest=RandomForestRegressor()
forest.fit(x_train_scaled, y_train)
y_forest=forest.predict(x_test) 
```

```python
y_rf_prob= forest.predict_proba(x_test)[:,1]
```

```python
#Performance metrics evaluation
print("Accuracy:\n",accuracy_score(y_test,y_tree))
print("Precision:\n",classification_report(y_test,y_tree))
```

```python
#Performance metrics evaluation
print("Accuracy:\n",metrics.accuracy_score(y_test,y_forest))
print("Precision:\n",classification_report(y_test,y_forest))

#plotting the ROC curve
fpr,tpr,thresholds=metrics.roc_curve(y_test,y_rf_prob)
plt.plot(fpr,tpr,'b', label='AUC = %0.2f'% auc)
plt.plot([0,1],[0,1],'r-.')
plt.xlim([-0.2,1.2])
plt.ylim([-0.2,1.2])
plt.title('Receiver Operating Characteristic\nRandom Forest')
plt.legend(loc='lower right')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
